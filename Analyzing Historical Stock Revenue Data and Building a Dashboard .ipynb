## Question 2: Use webscraping to Extract Tesla Revenue Data
Display the last five rows of the tesla_revenue dataframe using the tail function. Upload a screenshot of the results.


```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Define the URL of the webpage you want to download
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/revenue.htm"

# Send an HTTP GET request to the URL and fetch the HTML content
response = requests.get(url)
html_data = response.text

# Parse the HTML data with Beautiful Soup
soup = BeautifulSoup(html_data, 'html.parser')

# Find the table that contains the revenue data (you may need to inspect the HTML to find the exact structure)
table = soup.find('table')

# Create empty lists to store the data
dates = []
revenues = []

# Loop through the rows of the table (skip the header row)
for row in table.find_all('tr')[1:]:
    cols = row.find_all('td')
    if len(cols) == 2:
        date, revenue = [col.get_text(strip=True) for col in cols]
        dates.append(date)
        revenues.append(revenue)

# Create a DataFrame from the extracted data
tesla_revenue = pd.DataFrame({'Date': dates, 'Revenue': revenues})

# Display the first few rows of the tesla_revenue DataFrame
print(tesla_revenue.head())
```

```python
print(tesla_revenue.tail())
```

## Question 5: Plot Tesla Stock Graph
Use the make_graph function to graph the Tesla Stock Data, also provide a title for the graph.
Upload a screenshot of your results.

```python
import matplotlib.pyplot as plt

def make_graph(data_frame, x_col, y_col, title ):
    plt.figure(figsize=(12, 6))
    plt.plot(data_frame[x_col], data_frame[y_col], marker='o', linestyle='-', color='b')
    plt.title(title)
    plt.xlabel(x_col)
    plt.ylabel(y_col)
    plt.grid(True)
    plt.show()

# Call the make_graph function with appropriate parameters
make_graph(tesla_data, 'Date', 'Close', 'Tesla Stock Price Over Time')
```
